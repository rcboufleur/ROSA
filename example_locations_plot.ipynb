```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
from cartopy.feature import NaturalEarthFeature, LAND, COASTLINE
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import os 
from matplotlib.image import imread

def brazil_states(projection=ccrs.PlateCarree()):
    fname = os.path.join('NE1_50M_SR_W', 'NE1_50M_SR_W.tif')
    fig, ax = plt.subplots(figsize=(16, 12), subplot_kw=dict(projection=projection))
    ax.set_extent([-90, -30, -45, 20])
    # ax.stock_img()
    ax.imshow(imread(fname), origin='upper', transform=ccrs.PlateCarree(), extent=[-180, 180, -90, 90])
#     ax.add_feature(LAND)
    ax.add_feature(COASTLINE, edgecolor='dimgray')
    gl = ax.gridlines(draw_labels=True)
    gl.xlabels_top = True
    gl.ylabels_right = True
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    return fig, ax
```

```python
import numpy as np
array = np.array([[-36,-36.52],
                  [-6.2,-6.5]
                  []
        ])
array[1]
```

```python
fig, ax = brazil_states()
states = NaturalEarthFeature(category='cultural', scale='50m', facecolor='none',
                             name='admin_1_states_provinces_shp')
_ = ax.add_feature(states, edgecolor='dimgray', ls='--')
plt.scatter(-43.2096, -22.9035, marker='.', s=100, color='red')
plt.scatter(array[0],array[1], marker='*', s=50, color='black')
plt.annotate('LIneA/ON (Rio de Janeiro)', (-43.2096+0.5 , -22.9035+0.5), color='red', size=12)
```
