```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
from cartopy.feature import NaturalEarthFeature, LAND, COASTLINE
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import os 
from matplotlib.image import imread

def brazil_states(projection=ccrs.PlateCarree()):
    fname = os.path.join('NE1_50M_SR_W', 'NE1_50M_SR_W.tif')
    fig, ax = plt.subplots(figsize=(16, 12), subplot_kw=dict(projection=projection))
    ax.set_extent([-90, -30, -45, 20])
    ax.set_extent([-76, -34, -36, 10])

    # ax.stock_img()
    ax.imshow(imread(fname), origin='upper', transform=ccrs.PlateCarree(), extent=[-180, 180, -90, 90])
#     ax.add_feature(LAND)
    ax.add_feature(COASTLINE, edgecolor='dimgray')
    gl = ax.gridlines(draw_labels=True)
    gl.xlabels_top = True
    gl.ylabels_right = True
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    return fig, ax
```

```python
import numpy as np

# fit
x0 = [-55,-50, -42, -40]
y0 = [-30,-25, -19, -3]
c = np.polyfit(x0,y0,1)

n = 49
x = np.linspace(-54.45,-37.5,n) 
# y = c[0]*x*x + c[1]*x + c[2]
y = c[0]*x + c[1]

y[0:10]
np.random.seed(4)
x += np.random.normal(0,1.5,n) - 0.2

fig, ax = brazil_states()
states = NaturalEarthFeature(category='cultural', scale='50m', facecolor='none',
                             name='admin_1_states_provinces_shp')
_ = ax.add_feature(states, edgecolor='dimgray', ls='--')
plt.scatter(-43.2096, -22.9035, marker='*', s=60, color='red')
plt.scatter(x, y, marker='$$', s=50, color='black')
plt.scatter(-44.9661952, -23.0731859, marker='o', color='black')
plt.annotate('LIneA/ON (Rio de Janeiro)', (-43.2096+0.5 , -22.9035+0.5), color='red', size=10)
plt.annotate('Cunha/SP', (-44.9661952+0.5 , -23.0731859-1.), color='black', size=10)
```
